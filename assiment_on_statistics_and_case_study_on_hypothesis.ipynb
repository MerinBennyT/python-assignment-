2. A company started to invest in digital marketing as a new
way of their product promotions. For that they collected
data and decided to carry out a study on it.

a. The company wishes to clarify whether there is any
increase in sales after stepping into digital marketing.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
sales=pd.read_csv("/content/Sales_add.csv")
```

```python
#H0:There is any increase in sales after stepping into digital marketing
#H1:there is no increase in sales after stepping into digital marketing

before = sales['Sales_before_digital_add(in $)'].values
before
after = sales['Sales_After_digital_add(in $)'].values
after
```

```python
observed = np.array([before, after])

chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)
print(chi2_stat, p_value, dof, expected)
```

```python
alpha=0.05
if p_value<alpha:
  print("reject H0: there is any increase in sales after stepping into digital marketing")
else:
    print("fail to reject H0: there is any increase in sales after stepping into digital marketing")
```
